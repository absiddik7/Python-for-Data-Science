```python
import pandas as pd
import numpy as np
import cufflinks as  cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
# connect java script to notebook
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
# Data 
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]}) # data frame using dictonary
```

```python
df2
```

```python
df.iplot()
```

```python
#scatter plot
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
#bar plot
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
#count values of each columns
df.count().iplot(kind='bar')
```

```python
#sum of the each coulmns
df.sum().iplot(kind='bar')
```

```python
#box plot
df.iplot(kind='box')
```

```python
# 3D surface plot
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[300,400,500,600,700]})
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
#histogram
df['A'].iplot(kind='hist',bins=40)
```

```python
df.iplot(kind='hist',bins=20)
```

```python
#spread plot
df[['A','B']].iplot(kind='spread')
```

```python
#bubble plot
df.iplot(kind='bubble',x='A',y='B',size='C')
```
