```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
"""
Create a figure object called fig using plt.figure()
Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
Plot (x,y) on that axes and set the labels and titles to match the plot below:

"""
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
"""
Create a figure object and put two axes on it, ax1 and 
ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.
"""
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])


```

```python
"""
Now plot (x,y) on both axes. And call your figure object to show it.
"""
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
"""
Create the plot below by adding two axes to a 
figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
"""
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
"""
Now use x,y, and z arrays to recreate the plot below. 
Notice the xlimits and y limits on the inserted plot:
"""
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(y,z)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
"""
Use plt.subplots(nrows=1, ncols=2) to create the plot below.
"""
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b',lw=3,ls='--')
axes[1].plot(y,z,'r',lw=3,ls='-')
```

```python
"""
resize the plot by adding the figsize() argument in plt.subplots()
"""
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
axes[0].plot(x,y,'b',lw=3,ls='-')
axes[1].plot(y,z,'r',lw=3,ls='--')
```
