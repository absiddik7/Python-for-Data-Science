```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#allow to print the plot in here
%matplotlib inline 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# Functional Method to Print Plot

```python
# Functional Method to print plot
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
# Multiplot
plt.subplot(1,2,1) # row = 1, column = 2
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Way to Print Plot


```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom, width, height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# set 2 figure in one canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom, width, height
axes2 = fig.add_axes([0.2,0.48,0.4,0.3]) # left,bottom, width, height

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLLER PLOT')
```

```python
#Multiplot
fig,axes = plt.subplots(nrows=1,ncols=4)

for current_ax in axes:
    current_ax.plot(x,y)
    
plt.tight_layout() #avoid overlaping plot
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(4,2)) # width, height

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,4)) # width, height

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Saving Figure

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

```python
fig.savefig('my_plot.png',dpi=200)
```

### Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# indicate the plot label
ax.plot(x,x**2,label='x Squared') 
ax.plot(x,x**3,label='x Cubed')

ax.legend(loc=0) #loc determine the legend locations on the figure, 0 = best,... 10 = center
```

# Plot Appearance


### Line Color

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#FF8C00') #HEX color
```

### Line Width

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#FF8C00', linewidth=4) #HEX color
```

```python
# line tranparency
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#FF8C00', lw=4,alpha=0.5) #HEX color
```

### Line Style

```python
"""
available options:
'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
"""
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green', lw=4,linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green', lw=4,linestyle='-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green', lw=4,linestyle=':')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green', lw=4, ls='dotted')
```
