```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Bar plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# Bar plot - estimator. Like groupby
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # np.std = standard deviation
```

```python
# Count Plot
sns.countplot(x='sex',data=tips)
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Box plot-hue
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#violin plot- hue and split
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#Strip plot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#Strip plot-hue and split
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True) #dodge is split
```

```python
#Swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Violinplot with Swarmplot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# FactorPlot
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
