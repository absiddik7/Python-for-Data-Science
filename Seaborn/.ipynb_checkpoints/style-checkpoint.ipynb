```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Background

```python
# set white background
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
# set white background 
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# set white background 
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
# set white background 
sns.set_style('dark')
sns.countplot(x='sex',data=tips)
```

```python
# set white background 
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

### Aspect and Height

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

### Scale and contex


```python
# make graph for using in poster
sns.set_context('poster',font_scale=1)
sns.countplot(x='sex',data=tips)
```

### Palette and color

```python
#get palette from https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
