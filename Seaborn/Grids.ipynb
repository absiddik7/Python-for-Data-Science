```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
#pairplot
sns.pairplot(iris)
```

```python
#pair Grid
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# specify the plot type using pair grid
g = sns.PairGrid(iris)
g.map_diag(sns.histplot) # diagonaly histplot
g.map_upper(plt.scatter) # upper side scatter plot
g.map_lower(sns.kdeplot) # lower side kdeplot
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# facet grid
t = sns.FacetGrid(data = tips,col='time',row='smoker')
t.map(sns.histplot,'total_bill',kde=True)
```

```python
# facet grid
t = sns.FacetGrid(data = tips,col='time',row='smoker')
t.map(plt.scatter,'total_bill','tip')
```
